```python
from bs4 import  BeautifulSoup
from bs4.element import Comment


def process_html(raw_html):
    '''
    Общая функция для обработки html
    
    '''
    soup = BeautifulSoup(raw_html, "html.parser")
    
    # заголовок
    title = html2text_bs_title(soup)
    
    # видимый текст
    visible_text = html2text_bs_soup(soup)
    
    return visible_text, title
    

def html2text_bs_soup(soup):
    """
    Тут производится извлечение текста из html
    """
    [s.extract() for s in soup(['script', 'style',  '[document]', 'head', 'title', 'meta', 'head'])]
    return soup.get_text()

def html2text_bs_visible(raw_html):
    from bs4 import BeautifulSoup
    """
    Тут производится извлечения из html текста, который видим пользователю
    """
    soup = BeautifulSoup(raw_html, "html.parser")    
    [s.extract() for s in soup(['style', 'script', '[document]', 'head', 'title','meta'])]
    return soup.get_text()


def html2text_bs_title(soup):
    '''
    Извлечение title
    '''
    title = soup.title
    if title != None:
        return title.string
    else:
        return None
```

```python
def easy_tokenizer(text):
    word = str()
    for symbol in text:
        if symbol.isalnum(): word += symbol
        elif word:
            yield word
            word = str()
    if word: yield word

PYMORPHY_CACHE = {}
MORPH = None

def get_lemmatizer():
    import pymorphy2
    global MORPH
    if MORPH is None: MORPH = pymorphy2.MorphAnalyzer()
    return MORPH


def pymorphy_tokenizer(text):
    global PYMORPHY_CACHE
    
    words = list(filter(lambda x:  x
                          not in stopwords_en or x not in stopwords_rus
                          or x not in symbols or len(x) >= 3,
                           easy_tokenizer(text)))
    
    for word in words:
        word_hash = hash(word)
        if word_hash not in PYMORPHY_CACHE:
            PYMORPHY_CACHE[word_hash] = get_lemmatizer().parse(word)[0].normal_form            
        yield PYMORPHY_CACHE[word_hash]
```

```python
def parse_doc(doc):
    text = doc.text
    
    text_parts = {}
    if doc.text is None:
        print('unable read document: ' + str(doc))
        return None

    if doc.type == 'html':
        vis_text, title = process_html(text)
        
    else:
        vis_text, title = text, None
    vis_text_lem = pymorphy_tokenizer(vis_text)
    if title != None:
        title = pymorphy_tokenizer(title)
        
    text_parts['text'] = vis_text_lem
    text_parts['title'] = title
    return text_parts
```

## Собираем статистику слов  по документам

```python
class Doc_Info():
    def __init__(self):
        self.text = {}
        self.pos_in_text = {}
        
        self.title = {}
        self.pos_in_title = {}
        
        self.len = 0
        self.id = None

    def count_word(self, word):
        in_title = 0
        in_text = 0
        if word in self.title:
            in_title = self.title[word]
        if word in self.text:
            in_text = self.text[word]
        return in_title, in_text
```

```python
def extract_doc_stat(doc_id):
    with open(processed_folder + str(doc_id) + '.txt','r', encoding='UTF-8') as f:
        cur_doc = f.read()

    doc_stat = Doc_Info()

    data = cur_doc.split('SSSSPLITTTT')

    title = data[0]
    if title == 'NOTITLE':
        title = ''
    if len(data) == 2:
        text = data[1]

        pos = 0
        pos_bi = 0

        title_tokens = word_tokenize(title)
        for word in title_tokens:
            if word in doc_stat.title:
                doc_stat.title[word] += 1
            else:
                doc_stat.title[word] = 1
            if word in doc_stat.pos_in_title:
                    doc_stat.pos_in_title[word].append(pos)
            else:
                doc_stat.pos_in_title[word] = [pos]
            pos += 1

        text_tokens = word_tokenize(text)        
        for word in text_tokens:
            if word in doc_stat.text:
                doc_stat.text[word] += 1
            else:
                doc_stat.text[word] = 1

            if word in doc_stat.pos_in_text:
                doc_stat.pos_in_text[word].append(pos)
            else:
                doc_stat.pos_in_text[word] = [pos]
                pos += 1

        doc_stat.id = doc_id
        doc_stat.len = pos

        with open(data_folder+'doc_stat/' + str(doc_id) + '.stat', 'wb') as file:
            dill.dump(doc_stat, file)
        file.close()
        
        
        return 1
    else:
        return None
```

## Модифицированная BM52

Попробуем реализовать небольшое изменение BM25, которое было представлено в статье "Алгоритм текстового ранжирова-ния Яндекса на РОМИП-2006":

$$Score = W_{single} + W_{pair} + k_1  W_{allwords} + k_2  W_{phrase} + k3  W_{halfWords} + W_{PRF}$$

где 
$$p = IDF$$



$$W_{single} = log(p) \frac{TF}{TF + k_1 + k_2* DocLength} + 0.3 \frac{Hdr}{Hdr+1}$$

$$W_{pair} = 0.3 (log(p_1) + log(p_2)) \frac{TF}{1+TF}$$

Я в данном ноутбуке немного редуцирую формулу до следующего вида:

$$Score = W_{single} + W_{pair} + k_1  W_{allwords} + k_2  W_{phrase}$$

```python
doc_counts = len(docs_processed)
def idf(word):
    df = 0
    if word in corpus_dict.keys():
        df = corpus_dict[word].df
    return (doc_counts - df + 0.5)/ (df + 0.5)
def log_idf(word):
    word_idf = idf(word)
    if word_idf == 0:
        return 0
    return np.log(word_idf)
```

### Учет встречаемости единичных слов

```python
def W_single(word, doc):
    in_title, in_text= doc.count_word(word)
    tf = in_title + in_text
    score = tf / (tf + K1 + K2 * doc.len)
    return log_idf(word) * (score + 1.5 * int(in_title > 0))
```

```python
def Score_W_single(words, doc):
    score = 0
    for w in words:
        score += W_single(w, doc)
    return score
```

### Учет встречаемости пар слов

```python
def W_pair(w1, w2, pos1, pos2):
    count_row = 0
    count_inverse = 0
    count_skip = 0
    for p1 in pos1:
        if (p1 + 1) in pos2:
            count_row += 1
        if (p1 - 1) in pos2:
            count_inverse += 1
        if (p1 + 2) in pos2:
            count_skip += 1
    return count_row, count_inverse, count_skip
```

```python
def get_pos(doc, word):
    pos_title = []
    pos_text = []
    if word in doc.pos_in_title.keys():
        pos_title = doc.pos_in_title[word]
    if word in doc.pos_in_text.keys(): 
        pos_text = doc.pos_in_text[word]
    return pos_title, pos_text


def Score_W_pairs(words, doc):
    score = 0
    for i in range(len(words) - 1):
        w1 = words[i]
        w2 = words[i + 1]
        

        in_text1, in_title1 = get_pos(doc, w1)
        
        pos1 = in_title1 + in_text1
        
        in_title2, in_text2 = get_pos(doc, w2)

        
        pos2 = in_title2 + in_text2

        count_row, count_inverse, count_skip = W_pair(w1, w2, pos1, pos2)
        tf = count_row * ROW + count_inverse * INVERSE + count_skip * SKIP
        score += (W_single(w1, doc) + W_single(w2, doc))*tf/(1 + tf)
    return score
```

###  Учет наличия всех слов в запросе

```python
def W_allwords(words, doc, med_idf):
    score = 0
    # number of terms that are missing in doc
    miss = 0
    for w in words:
        w_idf = W_single(w, doc)
        in_title, in_text = doc.count_word(w)
        if in_title == 0 and in_text ==0:
            if w_idf > med_idf:
                miss += 1
        score += w_idf
    score *= MISS_PENALTY ** miss
    return score
```

### Учет наличия фразы целиком

```python
def W_phrase(query, query_ext, doc_id):
    
    COEF_TITLE = 3
    COEF_ORIG = 1
    COEF_SYN = 0.5
    
    
    score = 0
    
    with open(processed_folder + str(doc_id) + '.txt','r', encoding='UTF-8') as f:
        cur_doc = f.read()

    data = cur_doc.split('SSSSPLITTTT')

    title = data[0]
    if title == 'NOTITLE':
        title = ''
    if len(data) == 2:
        text = data[1]
        #print(text)
        
        score += COEF_ORIG * text.count(query) / (text.count(query) + 1)
        score += COEF_SYN * text.count(query_ext) /(text.count(query_ext) + 1)
        
        if title != '':
            score +=  COEF_TITLE * COEF_ORIG * title.count(query) / (title.count(query) + 1)
            score +=  COEF_TITLE * COEF_SYN * title.count(query_ext) / (title.count(query_ext) + 1)
        return score * 0.1
    else:
        return 0
```

## Расчет весов для документов

```python
COEF_SINGLE = 1.0
COEF_EXT = 0.4
K1 = 1.0
K2 = 1.0 / 350

COEF_PAIRS = 0.6
ROW = 1.3
INVERSE = 0.7
SKIP = 0.5

COEF_ALL_WORDS = 0.2
MISS_PENALTY = 0.03
COEF_PHRASE = 0.1


DOC_FOLDER = data_folder+'doc_stat/'
#data_folder+'doc_stat/' + str(doc_id) + '.stat'

result_yandex_idf = {}

for q_num in tqdm(queries.keys()):
    result_yandex_idf[q_num] = dict()
    
    query_orig = queries[q_num].words_lem
    query_ext = queries[q_num].words_ext_lem
    #print(query_orig, query_ext)
    for doc_id in queries[q_num].doc_indices:
        score = 0
        if os.path.exists(DOC_FOLDER + str(doc_id) + '.stat'):
            doc = dill.load(open(DOC_FOLDER + str(doc_id) + '.stat', 'rb'))
            #print(doc)
            score += COEF_SINGLE * Score_W_single(query_orig, doc)
            score += COEF_SINGLE * COEF_PAIRS* Score_W_pairs(query_orig, doc)
            
            score += COEF_EXT * Score_W_single(query_ext, doc)
            score += COEF_EXT * COEF_PAIRS * Score_W_pairs(query_ext, doc)
            
            query_idfs = []
            for w in query_orig:
                query_idfs.append(log_idf(w))
            med_idf = np.median(np.array(query_idfs))
            
            score += COEF_ALL_WORDS * W_allwords(query_orig, doc, med_idf)
            
            score += COEF_PHRASE * W_phrase(' '.join(queries[q_num].words_lem), ' '.join(queries[q_num].words_ext_lem), doc_id)

            
            result_yandex_idf[q_num][doc_id] = score
```

```python
def create_sub_dict(result, queries):
    sub_dict = dict()
    for q_num in tqdm(queries.keys()):
        doc_ids = []
        scores = []
        for i in result[q_num]:
            doc_ids.append(i)
            scores.append(result[q_num][i])
        scores = np.array(scores)
        doc_ids = np.array(doc_ids)
        doc_ids = doc_ids[np.argsort(scores)[::-1][:10]]
        sub_dict[q_num] = doc_ids
    return sub_dict
```

```python
def write_table(filename, res_dict):
    with open(filename + '.txt', 'w') as f:
        f.write('QueryId,DocumentId' + '\n')
        for q_id in res_dict:
            for res in res_dict[q_id]:
                f.write(str(q_id) + ',' + str(res) + '\n')
```

Функции для расстояний между эмбеддингами

```python
def arccosine_sim(emb1, emb2):
    sim = 1 - np.arccos(
      sklearn.metrics.pairwise.cosine_similarity(emb1,
                                                 emb2))/np.pi
    return sim

def cosine_sim(emb1, emb2):
    sim = sklearn.metrics.pairwise.cosine_similarity(emb1,
                                                 emb2)
    return sim
```

```python
def create_res_emb(result):

    result_emb = {}

    for q_id in tqdm(result.keys()):
        result_emb[q_id] = {}
        idx = []
        values = []
        for i in result[q_id]:
            idx.append(i)
            values.append(result[q_id][i])

        values = np.array(values)
        idx = np.array(idx)
        idx = idx[np.argsort(values)[::-1]]

        doc_emb_dict = None
        emb_file = data_folder + 'doc_emb' + q_id + '.pkl'

        if os.path.exists(emb_file):
            doc_emb_dict = pickle.load(open(emb_file, 'rb'))


        for doc_id in tqdm(idx[:10]):

            if doc_emb_dict:
                sim_q_d = cosine_sim(queries_embed[q_id], doc_emb_dict[doc_id])[0][0]
            else:
                text = get_text(docs_texts[doc_id])
                if len(text) > 2000000:
                    text=text[:2000000]
                doc_emd = embed_text(text)


                sim_q_d = cosine_sim(queries_embed[q_id], doc_emd)[0][0]
                #sim_q_d_ext = cosine_sim(queries_sim_embed[q_id], doc_emd)[0][0]


            result_emb[q_id][doc_id] = sim_q_d
    return result_emb
```

```python
def create_prediction(result_simple, result_emb, k1, k2):
    sub = dict()
    for q_id in tqdm(result_emb.keys()):
        doc_ids = []
        emb_weights = []
        for i in result_emb[q_id]:
            doc_ids.append(i)
            emb_weights.append(result_emb[q_id][i])

        #print(doc_ids)
        #print(emb_weights)
        bm_weights = []  
        for i in doc_ids:
            bm_weights.append(result_simple[q_id][i])


        #print(bm_weights)
        doc_ids = np.array(doc_ids)
        emb_weights = np.array(emb_weights)
        bm_weights = np.array(bm_weights)

        final_weights =  k1 * bm_weights + k2 * emb_weights
        sub[q_id] = doc_ids[np.argsort(final_weights)[::-1]][:10]

    return sub
```
